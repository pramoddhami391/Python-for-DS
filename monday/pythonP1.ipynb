```python
#sum of n natural numbers

def sum(n):
    s=0
    for i in range(1,n+1):
        s=s+i
    return s

print(sum(5))
```

```python
#WAP to return sum,difference,product and quotient using singlr fumction
def calculate(a,b):
    s=a+b
    d=a-b
    p=a*b
    q=a/b
    return s,d,p,q
s,d,p,q=calculate(4,2)
print(s)
 
```

```python
#*args
def average(*args):
    print(args)
    s=0
    for i in args:
        s=s+i
    return s
print(average(1,6))
```

```python
#**args
def func(**kwargs):#keyword argument
    print(kwargs)
    for item in kwargs.items():
        print(item)
   
func(a=10,b=20)
```

```python
def myFunc(*args,**kwargs):
    print(args)#argumants
    print(kwargs)#dictionary with key value
myFunc(1,2,a=10,b=20)
```
