```python
st=set()

st={"Banana","Orange","Apple"}

for fruit in st:
    print(fruit)
print (len(st))
```

We can use for in loop innn set to iterate through its elements.
we can use len(set) method to find the length(no. of elements) of a set

```python
print(f"Does st contains Banana?","Banana" in st)#checking an item
print("Apple" in st)
```

Add multiple items using update().

```python
st.update(["Mango","Avocado","Pomogranate"])
print(st)
```

```python
#set can be updated by tuple
fruits=("Grapes","Pineapple")
st.update(fruits)
print(st)
```

removing item from a set:use set.remove("item")

```python
st.remove("Grapes")
st
```

The pop() methods remove a random item from a list and it returns the removed item.

```python
st.pop()
st
```

```python
st.pop()
st
```

If we want to clear or empty the set we use clear method.

```python
st.clear()
st
```

deleting a set:use del operator

```python
st={"Apple","Banana"}
del st
```

converting list into set:set(list)

```python
list=["Apple","Banana",4,6]
st=set(list)
st
```

joining two sets:use set1.union(set2)

```python
st1={"Apple","Banana"}
st2={1,2}
st3=st1.union(st2)
st3
```
