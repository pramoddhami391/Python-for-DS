```python
#dictionary example
dict={"a":1,"b":2,"name":"ram"}
for key in dict:#iterating over the key
    print(key)
```

```python
person={"name":"Pramod","age":20}
length=len(person)
print(length)
for key in person:
    print(person[key])#accessing the value
```

```python
print(person)
print(person["name"])
print(person["age"])
```

Accessing an item by key name raises an error if the key does not exist. To avoid this error first we have to check if a key exist or we can use the get method. The get method returns None, which is a NoneType object data type, if the key does not exist.

```python
print(person.get("name"))
print(person.get)
```
