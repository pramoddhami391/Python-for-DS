4.How do you handle division by zero in Python?
->By using exception handelling.
eg:


```python
#exception is an unexpected error
try:
    n1=int(input("Enter n1:"))
    n2=int(input("Enter n2:"))
    result=n1/n2
    print(result)

except ValueError:
    print("Enter any valid number.")

except ZeroDivisionError:
    print("You cannot divide by zero")

finally:
    print("This always runs")
```

15.What are *args and **kwargs?
->*args can take arguments whereas **kwargs can take keyword arguments.
eg: 

```python
def sum(*args):
    print(args)
    s=0
    for i in args:
        print(i)
        s=s+i
    return s
print(sum(1,6))
```

```python
#**kwargs
def func(**kwargs):
    print(kwargs)
    for item in kwargs.items():
        print(item)

func(a=1,b=2)
```
