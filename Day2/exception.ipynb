```python
#exception is an unexpected error
try:
    n1=int(input("Enter n1:"))
    n2=int(input("Enter n2:"))
    result=n1/n2
    print(result)

except ValueError:
    print("Enter any valid number.")

except ZeroDivisionError:
    print("You cannot divide by zero")

finally:
    print("This always runs")
```

collection module:
counter-counts how many times each item appears.


```python
from collections import Counter

nums=[1,2,2,3,3,4]
c=Counter(nums)
print(type(c))
print(c)
dict={"name":"Pramod" ,"name":"Shyam"}
d=Counter(dict)
print(d)
```

```python
from collections import defaultdict
aDict=defaultdict(int)
aDict["a"]+=1
print(aDict)
print(aDict["a"])
```

deque
->fast for adding/removing from both sides

```python
from collections import deque

q=deque([1,2,3])
# q.append(0)#appent at right
# q.pop()#pop form right
# q.popleft()#pop from left
# q.clear
q.remove(1)

print(q)
```
