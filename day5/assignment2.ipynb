1.Write a Python program to check if a given number is prime.

```python
import math
def check(n):
    if n<2:
            return "Not a prime number."
            #this reduces the time complexity by 100x [time complexity=0(n^0.5)]
    for i in range(2,int(math.sqrt(n))+1):#range can't take float
        if n%i==0:
            return "Not a prime number"
    return "is a prime number"

result=check(1)
result
```

2.Create a function that converts a list of temperatures from Celsius to Fahrenheit.

```python
def c_to_f(temperatures):
    ftemp_lst=[]
    for temp in temperatures:
        f=9/5*(temp)+32
        ftemp_lst.append(f)
    return ftemp_lst

ctemp_lst=[35,36,37,40]
ftemp_lst=c_to_f(ctemp_lst)
ftemp_lst
```

4.Implement a function to find the factorial of a number using recursion.

```python
def fact(n):
    if n<0:
        return "Negative integers do not have a defined factorial."

    if n==1 or n==0:
        return 1
    if n>1:
        return n*fact(n-1)

result=fact(5)
result
```

7.Write a program to check if a string is a palindrome.

```python

def check(string):
    new_str=string.lower()
    if new_str==string[::-1]:
        return "The string is palindrome"
    else:
        return "The string is not palindrome."

string=input("Enter any string to check if it's palindrome or not.")
check(string)
```

8.Implement a class Circle with a method to calculate the area and circumference.

```python
import math
class Circle:
    def __init__(self,radius):
        if radius<0:
            raise ValueError("Radius cannot be negative")
        self.radius=radius

    def calc_area(self):
        return math.pi*(self.radius**2)
    
    def calc_circumference(self):
        return 2*math.pi*(self.radius)
    

radius=int(input("Enter the radius of the circle:"))    
c1=Circle(radius)
print(f"area={c1.calc_area():.2f}")
print(f"circumference={c1.calc_circumference():.2f}")
```

13.Create a class Rectangle with methods to calculate area and perimeter.

```python
class Rectangle:
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth

    def area(self):
        return self.length*self.breadth
    
    def perimeter(self):
        return 2*(self.length+self.breadth)
    
r1=Rectangle(4,2)
print(r1.area())
print(r1.perimeter())
        
```

15.Implement a Python function that generates the Fibonacci sequence up to n terms

```python

def fibo(n):
    fibo_s=[0,1]#if you declare this as a global list then it will keep extending as we call this function ,but we only need as our intended terms.
    if n==0:
        return "Invalid term"
    if n==1:
        return [0]
    if n==2:
        return fibo_s
    for i in range(2,n):
       
        fibo_s.append(fibo_s[-1]+fibo_s[-2])
    return fibo_s

fibo(4)


```

18.Implement a class Employee with attributes name and salary and a method to apply a 10% raise.

```python
class Employee:
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary

    def inx_salary(self,percent):
        self.salary+=(percent/100)*self.salary
        return self.salary  #this updates the salary of the object
    
e1=Employee("Ram",100000)
e1.inx_salary(10)
        
```

19.Write a Python program to check if two strings are anagrams of each other.

```python
#understanding the logic
str1="race"
str2="care"
if len(str1)!=len(str2):
        print("Not anagrams of eachother")
else:
    for char in str1:
      if str1.count(char)!=str2.count(char):
        print("Not anagrams of eachother")
        break
    else:
        print("Are anagrams")
        
```
