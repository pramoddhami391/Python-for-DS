```python
class Animal:
    def __init__(self,name,age):#constructor
        self.name=name
        self.age=age
    
    def __str__(self):#print(a1)
        return f"{self.name} is {self.age} years old."

    def __repr__(self):#a1
        return f"Animal:{self.name}|Age:{self.age}."
    
    def speak(self):#member function a1.speak()
        print(f"{self.name} is mowing.")

a1=Animal("Cow",5)
#a1.speak()
#print(a1)
#a1
```

```python
class Point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    
         
    
    def __add__(self,other):
        new_x=self.x+other.x
        new_y=self.y+other.y
        return Point(new_x,new_y)
    
    def __str__(self):
        return f"({self.x}, {self.y})"
    
p1=Point(2,3)
p2=Point(4,5)
p3=p1+p2#__add_ function is invoked
print(p3)
```
