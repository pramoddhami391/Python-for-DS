```python
class Animal:
    def __init__(self,name):
      self.name=name
    
    def speak(self):
       print(f"{self.name} speaks!")

class Dog(Animal):
    def __init__(self, name,breed):
      super().__init__(name)
      self.breed=breed

    def speak(self):
       super().speak()
       print(f"{self.name} barks!")

d1=Dog("Lucy","Sephard")
d1.speak()
# a1=Animal("Lion")
# a1.speak()
```
