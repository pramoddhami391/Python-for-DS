```python
#iterator
aList=[1,2,3,4]
it=iter(aList)
while True:
    try:
        num=next(it)
        print(num)
    except StopIteration:
        break
#Generator:divide big task into small chunks and do small chunk task at a time
#It saves memory because it produces value only when needed.
def square(x):
    for i in range(x):
        yield i*i
        
print(square(5))
print(type(square(5)))
#decorator
def decorator(func):
        def wrapper():
            print ("Before")
            func()
            print("After")
        return wrapper
#call th decorator
@decorator
def greet():
    print("Hello")
greet()
```
